## 2 -  Colorier des pixels
### 2.1 - Colorier un pixel
On peut colorier chaque pixel d’une image `img` en utilisant l’instruction `img.putpixel()` en précisant :
- Ses coordonnées au format `(x,y)`
- Sa couleur au format `(R, V, B)`

Le code suivant permet de créer une image de dimension `50*50 pixels` avec un fond blanc, de colorier le pixel de coordonnées `(25,25)` en rouge et enfin d’afficher l’image.
<div style="padding:4px; width:75px; border:3px solid #1560BD; background-color:#74D0F1; border-radius:15px;">
<strong>Script 2 :</strong> 
</div>

```python
# cellule à compléter
import matplotlib.pyplot as plt
from PIL import Image
img=Image.new("RGB",(50,50),(255,255,255)) # création d'une image avec un fond blanc
img.putpixel((25,25),(255,0,0)) # création d'un pixel rouge 
plt.imshow(img)
plt.show()
```

La fonction `putpixel` vient de l'anglais _put pixel_ que l'on peut traduire par mettre (dans le) pixel. Ainsi, l'instruction `img.putpixel((25, 25), (255, 0, 0))` peut se traduire par _mettre dans le pixel de coordonnées (25,25), la couleur (255, 0, 0) (rouge)_.<br>

### 2.2 - Colorier plusieurs pixels
Modifiez un pixel c'est déjà bien, mais comment faire pour modifier plusieurs pixels ?<br>
La réponse est simple, nous allons utiliser des boucles **<span style='color:#FF0000'>For.</span>**<br>
La structure de la boucle for est la suivante :<br>
for i in range(a, b)) :<br>
    instruction1<br>
    instruction2<br>
      suite programme<br><br>
Nous avons ici une boucle où la **<span style='color:#FF0000'>variable i</span>** prendra toutes les valeurs entières comprises entre a et b (**<span style='color:#FF0000'>a inclus</span>** et **<span style='color:#FF0000'>b exclu</span>**).<br>
<div style="padding:4px; width:75px; border:3px solid #1560BD; background-color:#74D0F1; border-radius:15px;">
<strong>Script 3 :</strong> 
</div>

```python
# cellule à exécuter
import matplotlib.pyplot as plt
from PIL import Image
#Creation d'une image blanche RGB 32x32
img = Image.new("RGB",(32,32),"white")

# definition d'une couleur
rouge = (255, 0, 0)

#remplissage de l'image avec une boucle:
for i in range(0,32): # pour les valeurs de i compris entre 0 et 32 pixels
    img.putpixel((0, i), rouge) # mettre en rouge les pixels dont les coordonnées (x=0 et y variant de 0 à 31)

#Affichage de l'image
plt.imshow(img)
plt.show()
```

### 2.3 - Colorier une plage de pixels


Si on veut colorier une plage de pixels, il suffit de colorier tous les pixels de cette plage. Pour cela, il suffit de parcourir tous les pixels `(x,y)` à colorier avec une double boucle `for`,comme ci-dessous pour une plage  de dimension `20*30` pixels à partir du pixel ayant pour coordonnées `(50,50)`.


_**Q7.** Expliquer les instructions qui permettent de créer ce rectangle ?_
<div style="padding:4px; width:75px; border:3px solid #1560BD; background-color:#74D0F1; border-radius:15px;">

</div>
Colorier un rectangle

```python
# cellule à exécuter
import matplotlib.pyplot as plt
from PIL import Image
img=Image.new("RGB",(100,100),"blue") # création d'une image avec un fond bleu
col=(255,0,0) # définition de la couleur utilisée
for x in range (50,70): # pour x compris entre 50 et 70
    for y in range (50,80): # pour y compris entre 50 et 80
        img.putpixel((x,y),col) # mettre en couleur les pixels x et y définis dans la double boucle 
plt.imshow(img)
plt.show()
```

Dans cette partie vous allez manipuler des images et changer les valeurs RGB des différents pixels.
Script 6 à compléter Affiche l'image initiale et les informations demandées

```python
# cellule à exécuter
from PIL import Image  # on importe le module de traitement d'images PIL (pillow)

img=Image.open("Ecureuil.jpg")  # on ouvre l'image initiale
r, v, b = img.getpixel((270, 230)) # on récupère la couleur des canaux rouge,vert,bleu du pixel situé aux coordonnées (90,150)
largeur, hauteur = img.size # on récupère la largeur et la hauteur de l'image en pixel
print("format =",img.format) # on affiche le format d'enregistrement de l'image
print("mode =",img.mode) # on affiche le mode de l'image
print("largeur =", largeur,"hauteur =", hauteur) # on affiche les dimensions
print("canal rouge : ", r, "canal vert : ", v, "canal bleu : ", b ) # on affiche la couleur des canaux
img.show() # On affiche l'image
```

Conversion de l'image en niveau de gris première méthode

```python
# cellule à exécuter
from PIL import Image  # on importe le module de traitement d'images PIL (pillow)

img=Image.open("Ecureuil.jpg")  # on ouvre le fichier image initial
largeur, hauteur = img.size
new_size = (largeur,hauteur)
img_ndg = Image.new('L',new_size)  # on crée une nouvelle image en niveaux de gris "img_ndg" de même taille
                    # le parametre 'L' indique une profondeur de couleur de 8
# boucles imbriquées permettant de parcourir tous les pixels 
for x in range(largeur):     
    for y in range(hauteur):
        r,v,b=img.getpixel((x,y))   # on récupère les 3 composantes de chaque pixel parcouru 
        ndg = int((r+v+b)/3)     # on calcul la valeur entière(int) de chaque pixel que l'on stocke dans la variable "ndg"
        img_ndg.putpixel((x,y),ndg) # on affecte à chaque pixel parcouru la valeur de la variable "ndg" à la nouvvelle image
img_ndg.show() # on affiche l'image modifiée
```

Compléter le Script pour réaliser le négatif de l'image
